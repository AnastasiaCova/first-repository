```python
x = int(input('Введите координаты x: '))
y = int(input('Введите координаты y: '))
x1=0 
y1=0
steps = 0
komp = 0
m = None #переменная, куда будет записываться минимальное количество шагов
k = 'север'
right = {'север':'восток', 'восток':'юг', 'юг':'запад', 'запад':'север'} #словари, по которым меняется направление 
left = {'север':'запад', 'запад':'юг', 'юг':'восток', 'восток':'север'}
reverse = {'север':'юг', 'юг':'север', 'восток':'запад', 'запад':'восток'}
n = input('Введите направление: ')
while n!='стоп': #пока не напишем стоп
    steps +=1 
    if n == 'вперед':   #когда идем вперед, записываем количество шагов на x или y, в зависимости от направления
        if k == 'север':
            y1 += int(input())
        elif k == 'запад':
            x1 -= int(input())
        elif k == 'юг':
            y1 -= int(input())
        elif k == 'восток':
            x1 += int(input())
    elif n == 'направо':
        k = right[k] #меняем направление
    elif n == 'налево':
        k = left[k]
    elif n == 'разворот':
        k = reverse[k]
    if x == x1 and y == y1 and not m: #когда координаты клада станут равны x1 и y1, то в переменную m запишется минимальное количество шагов
        m = steps
        komp = k
    n = input()
else:
    print('Минимальное количество шагов: ', m)
    print('Направление: ', komp)
        
```
